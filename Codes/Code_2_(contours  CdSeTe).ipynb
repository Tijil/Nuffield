```python
%pylab inline
import csv
```

```python
dataPath = 'C:/Users/tijil/Downloads/Nuffield-master (1)/Nuffield-master/CdSeTe/ellipsometery/CdSeTe/160701/elipss.txt'
x = []
y = []
z = []
```

```python
with open(dataPath) as f:
    data = csv.reader(f, delimiter='\t')
    for item in data:
        x.append(-float(item[1]))
        y.append(float(item[0]))
        z.append(float(item[2])/10)
```

```python
res =  20
xi = linspace(min(x), max(x), res) 
yi = linspace(min(y), max(y), res)
```

```python
zi = griddata(x, y, z, xi, yi, interp='linear')
```

```python
fig = figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.set_aspect('equal')
S = ax.contour(xi, yi, zi, 10, colors='k') 
C = ax.contourf(xi, yi, zi, 100, cmap='jet')
plt.xlabel('x position', fontsize=20)
plt.ylabel('y position', fontsize=20)
cb = fig.colorbar(C)
cb.set_label(r'Thickness $(nm)$', rotation=90, fontsize=20)
matplotlib.rc('xtick', labelsize= 20)
matplotlib.rc('ytick', labelsize= 20)
figtext(.1, -.04, "\nFigure 9: Data on how thickness of CdTe changes across the substrate. 317 points were collected \n to get this information via ellipsometry and the gaps were filled in by interpolation to make a contour graph.")
```

```python
dataPath_3 = 'C:/Users/tijil/Downloads/Nuffield-master (1)/Nuffield-master/CdSeTe/ellipsometery/CdSeTe/160701/elips2.txt'
x_2 = []
y_2 = []
z_2 = []
comp = []
n = 0
```

```python
with open(dataPath_3) as f:
    data = csv.reader(f, delimiter='\t')
    for item in data:
        x_2.append(float(item[0]))
        y_2.append(float(item[1]))
        z_2.append(float(item[2])/10)
```

```python
res =  20
xi = linspace(min(x), max(x), res) 
yi = linspace(min(y), max(y), res)
```

```python
zi = griddata(x_2, y_2, z_2, xi, yi, interp='linear')
```

```python
fig = figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.set_aspect('equal')
S = ax.contour(xi, yi, zi, 10, colors='k') 
C = ax.contourf(xi, yi, zi, 100, cmap='jet')
plt.xlabel('x position', fontsize=20)
plt.ylabel('y position', fontsize=20)
cb = fig.colorbar(C)
cb.set_label(r'Thickness $(nm)$', rotation=90, fontsize=20)
matplotlib.rc('xtick', labelsize= 20)
matplotlib.rc('ytick', labelsize= 20)
figtext(.1, -.04, "\nFigure 10: Data on how thickness of CdSe changes across the substrate. 317 points were collected \n to get this information via ellipsometry and the gaps were filled in by interpolation to make a contour graph.")
```

```python
for item in z:
    comp.append(float((z_2[n])/(item+z_2[n])))
    n += 1
```

```python
zi_comp = griddata(x, y, comp, xi, yi, interp='linear')
```

```python
fig = figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.set_aspect('equal')
S = ax.contour(xi, yi, (zi_comp*10), 10, colors='k') 
C = ax.contourf(xi, yi, (zi_comp*10), 100, cmap='jet')
plt.xlabel('x position', fontsize=20)
plt.ylabel('y position', fontsize=20)
cb = fig.colorbar(C)
cb.set_label('Composition (%Se)', rotation=90, fontsize=20)
matplotlib.rc('xtick', labelsize= 20)
matplotlib.rc('ytick', labelsize= 20)
figtext(.1, -.04, "\nFigure 11: Data on how composition of CdSe to CdTe changes across the substrate. 317 points were collected \n to get this information via ellipsometry and the gaps were filled in by interpolation to make a contour graph.")
```
