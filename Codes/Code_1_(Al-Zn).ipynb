```python
%pylab inline
import json
```

```python
dataPath = 'C:/Users/tijil/Downloads/datalogging/TCO_data.json'
```

```python
with open(dataPath) as f:
    data = json.load(f)
```

```python
d = []
x = []
y = []
Rs = []
mob = []
ne = []
for item in data:
    d.append(data.get(item).get('d'))
    x.append(data.get(item).get('x'))
    y.append(data.get(item).get('y'))
    Rs.append(data.get(item).get('Rs'))
    mob.append(data.get(item).get('mob'))
    ne.append(data.get(item).get('ne'))
```

```python
res =  1000 
xi = linspace(min(x), max(x), res) 
yi = linspace(min(y), max(y), res)
```

```python
zi = griddata(x, y, d, xi, yi, interp='linear')
```

```python
fig = figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.set_aspect('equal')
S = ax.contour(xi, yi, zi, 10, colors='k') 
C = ax.contourf(xi, yi, zi, 100, cmap='jet')
plt.xlabel('x position', fontsize=20)
plt.ylabel('y position', fontsize=20)
cb = fig.colorbar(C)
cb.set_label(r'Thickness $(nm)$', rotation=90, fontsize=20)
matplotlib.rc('xtick', labelsize= 20)
matplotlib.rc('ytick', labelsize= 20)
figtext(.1, -.04, "\nFigure 1: Contour plot showing the change thickness across the film, extracted from \n transmittance spectra, with respect to position over at 10 x 10 cm2 AZO film on a soda-lime \n glasss substrate.")
```
