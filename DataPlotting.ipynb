```python
Rs = []
d = []
mob = []
ne = []
rho = []
wp = []
x = []
y = []
```

```python
Rs
```

```python
for item in data:
    Rs.append(data.get(item).get('Rs'))
for item in data:
    d.append(data.get(item).get('d'))
for item in data:
    mob.append(data.get(item).get('mob'))
for item in data:
    ne.append(data.get(item).get('ne'))
for item in data:
    rho.append(data.get(item).get('rho'))
for item in data:
    wp.append(data.get(item).get('wp'))
for item in data:
    x.append(data.get(item).get('x'))
for item in data:
    y.append(data.get(item).get('y'))
```

```python
Rs = np.reshape(Rs, (50, 2))
```

```python
Rs
```

```python
xlist = np.linspace(-45, 45, 2)
ylist = np.linspace(-45, 45, 50)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(Rs)
plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.title('Rs')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```
